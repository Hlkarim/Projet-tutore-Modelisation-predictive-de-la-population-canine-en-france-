```python
import joblib
import pandas as pd
import geopandas as gpd

#from scripts.Carto.Carto_n_graphs import gdf_nan
```

```python
df25 = pd.read_csv('../../data/pop_humain_2025')
```

```python
gdf = gpd.read_file('../../Outputs/data/complete_output_lagged_V3.gpkg')
```

```python
gdf
```

```python
gdf25 = gdf.merge(df25, on='INSEE_COM', how='inner')
```

```python
gdf25
```

```python
gdf25.columns
```

```python
# 1. Charger le modèle
model = joblib.load('../../Outputs/modele/rf_model_imputed_temporal_small.joblib')

# --- 1. Préparation des données 2025 ---
# On fait une copie pour ne pas casser l'original
X_2025_ready = gdf25.copy()

# --- 2. LE SWAP (Crucial) ---
# On prend les chiffres de 2025, mais on les met dans la colonne "Pop_huma20"
# car c'est le seul nom que le modèle connait.
#X_2025_ready['Pop_huma20'] = X_2025_ready['Pop_huma13']

# --- 3. LE FILTRAGE STRICT ---
# On ne garde que les colonnes que le modèle attend, dans le bon ordre.
# rf_final.feature_names_in_ contient la liste exacte ['Pop_huma20', 'densite_pop', etc.]
features_attendues = model.feature_names_in_
X_final = X_2025_ready[features_attendues].copy()

# --- 4. Nettoyage final (Sécurité) ---
X_final = X_final.fillna(X_final.median())

# --- 5. Prédiction ---
print(f"Colonnes envoyées au modèle : {list(X_final.columns)}")
# Doit afficher exactement : ['Pop_huma20', 'densite_pop', 'Y_CHF_LIEU', 'Z_MOYEN', 'ratio_lag_final', 'SUPERFICIE_calc']

predictions_2025 = model.predict(X_final)

print(f"Total estimé 2025 : {predictions_2025.sum():,.0f}")
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# --- 1. Réintégration des prédictions dans le DataFrame maître ---
# On utilise data_2025 car c'est lui qui contient 'NOM_DEPT'
# (X_final ne contenait que les variables mathématiques)
df_resultats = gdf25.copy()
df_resultats['Chiens_Predits_2025'] = predictions_2025

# --- 2. L'Agrégation (Group By) ---
# On groupe par département et on somme la colonne des chiens
synthese_dept = df_resultats.groupby('NOM_DEPT')['Chiens_Predits_2025'].sum().reset_index()

# --- 3. Mise en forme et Classement ---
# On trie du plus grand au plus petit pour voir les "gros" départements
synthese_dept = synthese_dept.sort_values(by='Chiens_Predits_2025', ascending=False)

# On arrondit pour faire propre (pas de demi-chien)
synthese_dept['Chiens_Predits_2025'] = synthese_dept['Chiens_Predits_2025'].round(0).astype(int)

# --- 4. Affichage du TOP 10 ---
print("TOP 10 des Départements (Population Canine 2025 estimée) :")
print("-" * 60)
print(synthese_dept.head(10).to_string(index=False))
print("-" * 60)

# --- 5. Export Excel / CSV pour ton mémoire ---
synthese_dept.to_csv('projection_chiens_par_departement_2025.csv', index=False)
print("Fichier 'projection_chiens_par_departement_2025.csv' généré avec succès.")

# --- 6. Visualisation Graphique ---
plt.figure(figsize=(12, 8))
# On prend les 20 premiers pour que le graph reste lisible
top_20 = synthese_dept.head(20)

sns.barplot(x='Chiens_Predits_2025', y='NOM_DEPT', data=top_20, palette='viridis')
plt.title('Top 20 Départements : Projection Population Canine 2025', fontsize=14)
plt.xlabel('Nombre de chiens estimés')
plt.ylabel('Département')
plt.grid(axis='x', linestyle='--', alpha=0.6)
plt.tight_layout()
plt.show()
```
