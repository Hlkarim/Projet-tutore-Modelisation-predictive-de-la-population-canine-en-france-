# Historical Temporal imputation

```python
import numpy as np

# --- ETAPE 1 : Préparation des données ---
# On s'assure que toutes les colonnes de population sont numériques
years = range(13, 21)
for year in years:
    if f'Pop_huma{year}' in gdf.columns:
        gdf[f'Pop_huma{year}'] = pd.to_numeric(gdf[f'Pop_huma{year}'], errors='coerce')

    # Calcul des ratios pour la validation (étape indispensable pour savoir si la donnée est bonne)
    gdf[f'ratio_exact_{year}'] = gdf[f'Po_chien{year}'] / gdf[f'Pop_huma{year}'].replace(0, np.nan)

```

```python
# --- ETAPE 2 : Définition de l'Intervalle de Validité (Référence 2020) ---
mask_reference = gdf['typologie_normative'] == '0_Conforme_Standard'

dept_stats = gdf[mask_reference].groupby('NOM_DEPT')[f'ratio_exact_20'].agg(
    Q1=lambda x: x.quantile(0.25),
    Q3=lambda x: x.quantile(0.75)
).reset_index()

dept_stats['IQR'] = dept_stats['Q3'] - dept_stats['Q1']
dept_stats['Lower_Bound'] = (dept_stats['Q1'] - 1.5 * dept_stats['IQR']).clip(lower=0)
dept_stats['Upper_Bound'] = dept_stats['Q3'] + 1.5 * dept_stats['IQR']

# Fusion des bornes
cols_to_drop = [c for c in ['Lower_Bound', 'Upper_Bound'] if c in gdf.columns]
gdf = gdf.drop(columns=cols_to_drop)
gdf = gdf.merge(dept_stats[['NOM_DEPT', 'Lower_Bound', 'Upper_Bound']], on='NOM_DEPT', how='left')
```

```python
# --- ETAPE 3 : La Fonction Hybride (Croissance Démographique) ---

def apply_demographic_growth_robust(row):
    if pd.isna(row['Lower_Bound']):
        return row['Po_chien20']

    for year in range(20, 12, -1):
        chiens_hist = row[f'Po_chien{year}']
        ratio_hist = row[f'ratio_exact_{year}']
        pop_hist = row[f'Pop_huma{year}']
        pop_2020 = row['Pop_huma20']

        # 1. Vérifier si on a au moins le nombre de chiens historique et son ratio
        if pd.isna(chiens_hist) or pd.isna(ratio_hist):
            continue

        # 2. VALIDATION : Le ratio était-il cohérent à l'époque ?
        if row['Lower_Bound'] <= ratio_hist <= row['Upper_Bound']:

            # --- CAS A : On a les deux populations (Calcul du Growth Rate) ---
            if pd.notna(pop_hist) and pop_hist > 0 and pd.notna(pop_2020):
                growth_rate = pop_2020 / pop_hist
                return chiens_hist * growth_rate

            # --- CAS B : La population historique est manquante (Fusion de communes) ---
            # On ne peut pas calculer la croissance, donc on applique le ratio historique
            # directement à la population 2020 (qui, elle, est souvent présente).
            elif pd.notna(pop_2020):
                return ratio_hist * pop_2020

            # --- CAS C : Vraiment aucune population 2020 ---
            # On garde la valeur brute historique par défaut
            else:
                return chiens_hist

    return np.nan
```

```python
# --- ETAPE 4 : Exécution ---
print("Application du Taux de Croissance Humain sur les données historiques...")
gdf['Po_chien_history_growth'] = gdf.apply(apply_demographic_growth_robust, axis=1)

# Arrondi (on ne coupe pas les chiens en deux)
gdf['Po_chien_history_growth'] = gdf['Po_chien_history_growth'].round(0)
```

```python
# --- BILAN ---
nb_valide = gdf['Po_chien_history_growth'].notna().sum()
nb_correction = (
    (gdf['Po_chien20'].notna()) &
    (gdf['Po_chien_history_growth'].notna()) &
    (gdf['Po_chien20'] != gdf['Po_chien_history_growth'])
).sum()

print(f"Total communes validées : {nb_valide}")
print(f"Nombre de corrections apportées à 2020 : {nb_correction}")
```
