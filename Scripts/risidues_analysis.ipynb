```python
import matplotlib.pyplot as plt
import seaborn as sns

df_clean = gdf.copy()

# 1. Inspecter les extrêmes (pour comprendre l'ampleur du problème)
print("Statistiques des résidus relatifs (%) :")
print(df_clean['residu_relatif'].describe())

# 2. Filtrer pour le graphique (On garde ce qui est entre -200% et +200%)
# Cela couvre la majorité des cas "normaux"
subset_viz = df_clean[df_clean['residu_relatif'].between(-200, 200)]

plt.figure(figsize=(12, 6))

# 3. Augmenter les bins (50 ou 100 pour avoir de la finesse) et ajouter la courbe (KDE)
sns.histplot(subset_viz['residu_relatif'], bins=100, kde=True, color='teal')

plt.title("Distribution des Résidus Relatifs (Zoom sur +/- 200%)")
plt.xlabel("Erreur en % (0 = Prédiction Parfaite)")
plt.ylabel("Nombre de Communes")

# Ligne zéro (Perfection)
plt.axvline(x=0, color='red', linestyle='--', linewidth=2, label="Prédiction Parfaite")
plt.legend()

plt.show()

# 4. Compter combien de communes on a caché (les outliers)
n_outliers = len(df_clean) - len(subset_viz)
print(f"Attention : {n_outliers} communes (outliers extrêmes) sont masquées sur ce graphique.")
```

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import libpysal
from esda.moran import Moran, Moran_Local
from matplotlib.colors import ListedColormap, BoundaryNorm
import matplotlib.patches as mpatches

# --- 1. Préparation des Données (Sécurisée) ---
data_lisa = gdf.dropna(subset=['residu_relatif', 'geometry']).copy()
data_lisa = data_lisa[data_lisa.geometry.notna() & (~data_lisa.geometry.is_empty)]
data_lisa = data_lisa[~data_lisa['residu_relatif'].isin([np.inf, -np.inf])]
data_lisa = data_lisa[data_lisa['residu_relatif'].between(-500, 500)]

# --- 2. Calcul des Centroïdes ---
centroids = data_lisa.geometry.centroid
mask_coords_finies = np.isfinite(centroids.x) & np.isfinite(centroids.y)
data_lisa = data_lisa[mask_coords_finies]
centroids = centroids[mask_coords_finies]
coordinates = np.column_stack((centroids.x, centroids.y))

# --- 3. Matrice de Poids ---
w = libpysal.weights.KNN.from_array(coordinates, k=8, ids=data_lisa.index.tolist())
w.transform = 'r'

# --- 4. Moran Global et Local ---
y = data_lisa['residu_relatif'].values
moran = Moran(y, w)
lisa = Moran_Local(y, w)

# --- 5. Visualisation LISA (Filtre HH uniquement) ---
data_lisa['lisa_q'] = lisa.q
data_lisa['lisa_p'] = lisa.p_sim

# LOGIQUE DE FILTRAGE :
# On ne garde que les HH (q=1) ET significatifs (p <= 0.05)
# Tout le reste (LL, LH, HL ou non significatif) est mis à 0
mask_hh_sig = (data_lisa['lisa_p'] <= 0.05) & (data_lisa['lisa_q'] == 1)
data_lisa['lisa_visualisation'] = np.where(mask_hh_sig, 1, 0)

fig, ax = plt.subplots(figsize=(12, 12))

# Palette simplifiée : 0 = Gris (Fond), 1 = Rouge (HH)
colors = ['lightgrey', '#d7191c']
cmap = ListedColormap(colors)
norm = BoundaryNorm([-0.5, 0.5, 1.5], cmap.N)

data_lisa.plot(column='lisa_visualisation', cmap=cmap, norm=norm, ax=ax)

# Légende épurée
legend_labels = [
    mpatches.Patch(color='#d7191c', label='HH : Zones de Sous-estimation significative (Modèle trop bas)'),
    mpatches.Patch(color='lightgrey', label='Autres (Non significatif ou autres types de clusters)')
]
ax.legend(handles=legend_labels, loc='lower left', fontsize=10)

ax.set_axis_off()
plt.title("Focus sur les Clusters HH (Sous-estimation du modèle)\nSeulement les zones où p < 0.05")
plt.show()
```
